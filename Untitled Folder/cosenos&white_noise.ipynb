# Ejemplo de FFT $\left(\sum Cosenos + White Noise \right)$

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.fftpack import fft, ifft, fftshift, fftfreq

%matplotlib inline
```

## Definición de los cosenos
Creamos uno coseno (fundamental) de frecuencia $f_0$, y 3 armónicos.

Se define a los [armónicos](https://en.wikipedia.org/wiki/Harmonic "Harmonics Wiki") como lso cosenos cuya frecuencia son múltiplos de la frecuencia fundamental.

$$H_n = A_{[n]} \cdot  \cos (n \omega_0 t)$$

donde:

$$H_n \: \text{es el armónico y} \: \omega_0 = 2\pi f_0$$



```python
# Fundamental o Primer armónico
A = 1/3
f0 = 100
phi0 = 0


# Segundo armónico
A1 = 1/2
f1 = f0*2
phi1 = 0


# Tercer armónico
A2 = 1/1.5
f2 = f0*3
phi2 = 0


# Cuarto armónico
A3 = 1
f3 = f0*4
phi3 = 0

```

## Muestreo
Para poder crear las señales, debido a que nos encontramos en el campo discreto, necesitamos definir el espacio temporal. Eso se traduce en definir la _**ventana de tiempo**_ $t=[t_min, t_max]$ que queremos visualizar y la **_frecuencia de muestreo_** $f_s$.

Entonces el _step_ temeporal va a estar dado por:
$$t_s = \frac{1}{f_s}$$

Debe recordarse el teorema de [_Nyquist-Shannon_](https://en.wikipedia.org/wiki/Nyquist%E2%80%93Shannon_sampling_theorem "Nyquist Theorem Wiki"), que especifíca que:
$$f_s \ge 2f_{max}$$

```python
# Seteamos la fs 20 veces más que la frecuencia máxima de la señal. Esto nos dará una buena resolución temporal
fs = f3*20 
ts = 1/fs

# Seteamos los valores máximos y mínimos de la ventana temporal para que se puedan ver 3 ciclos
# de la señal de frecuencia más baja
cycles = 5

# Creamos la variable temporal
t = np.arange(0, cycles/f0, 1/fs)
```

## Creación de los cosenos

Definido el espacio temporal, podemos ahora crear las funciones cosenoidales, como también la suma de todos ellos.

```python
# Fundamental o Primer armónico
x0 = A * np.cos( 2 * np.pi * f0 * t + phi0 )

# Segundo armónico
x1 = A1 * np.cos( 2 * np.pi * f1 * t + phi1 )

# Tercer armónico
x2 = A2 * np.cos( 2 * np.pi * f2 * t + phi2 )

# Cuarto armónico
x3 = A3 * np.cos( 2 * np.pi * f3 * t + phi3 )

# Resultante
x = x0+x1+x2+x3


plt.figure(figsize=(20,5))
plt.subplot(1,2,1)
pl_x0 = plt.plot(t,x0)
pl_x1 = plt.plot(t,x1)
pl_x2 = plt.plot(t,x2)
pl_x3 = plt.plot(t,x3)
plt.legend(['x0','x1','x2','x3'])
plt.title("Cosenos")
plt.xlabel("Tiempo [seg]")
plt.ylabel("Amplitud")


plt.subplot(1,2,2)
plt.plot(t,x)
plt.title("Suma de cosenos")
plt.xlabel("Tiempo [seg]")
plt.ylabel("Amplitud")
```

# Agregado de Ruido blanco

Para hacer las cosas más divertidas y más semejantes a la realidad, se le agrega a la señal [ruido blanco](https://en.wikipedia.org/wiki/White_noise "White Noise Wiki").

Este ruido, es una señal _estocástica_ con distribución de probabilidad **normal** y tiene la particularidad de tener un espectro idealmente _**plano**_. La fase no se muestra, ya que son valores totalmente aleatoreos.

```python
# Creamos al ruido
An = 1
white_noise = An * np.random.normal(0, 1, len(t))

N = fftshift(fft(white_noise))
freq = fftshift(fftfreq(N.size, d=1/fs))
Nm = np.abs(N)/N.size

plt.figure(figsize=(20,5))

plt.subplot(1,2,1)
plt.plot(t, white_noise)
plt.title("Ruido Blanco")
plt.xlabel("Tiempo [seg]")
plt.ylabel("Amplitud")


plt.subplot(1,2,2)
plt.plot(freq, Nm)
plt.title("Módulo del Ruido Blanco")
plt.xlabel("Frecuencia [Hz]")
plt.ylabel("Amplitud")
```

Le sumamos entonces, el ruido a la señal

```python
xn = x + white_noise

plt.plot(t,xn)
plt.title("Cosenos + Ruido Blanco")
plt.xlabel("Tiempo [seg]")
plt.ylabel("Amplitud")
```

## Cálculo de la FFT total

Finalmente calculamos la Transformada de Fourier de la señal total.

**Nota:** típicamente todas las librerías que implementan la *FFT*, devuelven la zona de frecuencias negativas en la segunda mitad del array. Es por ello que se debe hacer un *shifteo* del array para poder visualizarlo correctamente.

**Nota:** para evitar ver saltos de fase en el gráfico (180° a -180°) se suele aplicar un [unwrap](https://en.wikipedia.org/wiki/Instantaneous_phase "Phase Unwrapping Wiki") a la fase.

```python
X = fftshift(fft(xn))
freq = fftshift(fftfreq(X.size, d=1/fs))

Xm = np.abs(X)/X.size
Xp = np.unwrap(np.angle(X))

plt.figure(figsize=(20,5))

plt.subplot(1,2,1)
plt.plot(freq, Xm)
plt.title("Modulo")
plt.xlabel("Frecuencia [Hz]")
plt.ylabel("Amplitud")
#plt.xlim([-1500,1500])

plt.subplot(1,2,2)
plt.plot(freq, Xp)
plt.title("Fase")
plt.xlabel("Frecuencia [Hz]")
plt.ylabel("Fase [Deg]")
#plt.xlim([-1500,1500])
```

### Módulo en decibeles

Por último visualizamos el espectro en decibeles, como suele hacerse

```python
Xm_dB = 20*np.log(np.abs(X)/np.max(np.abs(X)))

plt.figure()
plt.plot(freq, Xm_dB)
#plt.xlim([-2000,2000])
#plt.ylim([-70,0])
```
