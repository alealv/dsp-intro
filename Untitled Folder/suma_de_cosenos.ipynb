```python
%matplotlib inline

import numpy as np
import matplotlib.pyplot as plt
from scipy.fftpack import fft, ifft, fftshift, fftfreq
```

```python
A = 1

f0 = 100
f1 = f0*2
f2 = f0*3
f3 = f0*4


c = 10
fs = f3*20
t = np.arange(0,c/f0,1/fs)



x0 = A/4*np.cos(2*np.pi*f0*t)
x1 = A/3*np.cos(2*np.pi*f1*t)
x2 = A/2*np.cos(2*np.pi*f2*t)
x3 = A/1*np.cos(2*np.pi*f3*t)

x = x0+x1+x2+x3
```

```python
c2 = 3
subsample = int(c2/f0*fs)

#plt.plot(t[:subsample],x0[:subsample])
#plt.plot(t[:subsample],x1[:subsample])
#plt.plot(t[:subsample],x2[:subsample])
#plt.plot(t[:subsample],x3[:subsample])

plt.plot(t[:subsample],x[:subsample])
plt.title("Suma de cosenos")
plt.xlabel("Tiempo [seg]")
plt.ylabel("Amplitud")
```

```python
#noise = A*5 * np.random.rand(len(t))
noise = A * np.random.normal(0, 1, len(t))

c2 = 10
subsample = int(fs/f0*c2)

N = fftshift(fft(noise))
freq = fftshift(fftfreq(N.size, d=1/fs))
Nm = np.abs(N)/N.size



#plt.figure(figsize=(20,5))

#plt.subplot(1,2,1)
plt.plot(t[:subsample],noise[:subsample])
plt.title("Ruido Blanco")
plt.xlabel("Tiempo [seg]")
plt.ylabel("Amplitud")

#plt.subplot(1,2,2)
#plt.plot(freq, Nm)
```

```python
#xn = x + noise
xn = x


c2 = 10
subsample = int(fs/f0*c2)
plt.plot(t[:subsample],xn[:subsample])
plt.title("Dominio Temporal")
plt.xlabel("Tiempo [seg]")
plt.ylabel("Amplitud")
```

```python
X = fftshift(fft(xn))
freq = fftshift(fftfreq(X.size, d=1/fs))

Xm = np.abs(X)/X.size
Xp = np.unwrap(np.angle(X))

plt.figure()
plt.plot(freq, Xm)
plt.title("Modulo")
plt.xlabel("Frecuencia [Hz]")
plt.ylabel("Amplitud")
plt.xlim([-1500,1500])

plt.figure()
plt.plot(freq, Xp)
plt.title("Fase")
plt.xlabel("Frecuencia [Hz]")
plt.ylabel("Amplitud")
plt.xlim([-1500,1500])
```

```python
Xm_dB = 20*np.log(np.abs(X)/np.max(np.abs(X)))

plt.figure()
plt.plot(freq, Xm_dB)
#plt.xlim([-500,500])
plt.ylim([-70,0])
```
